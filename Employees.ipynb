### Overtime Hours Column

```python
df['Overtime Hours'].describe()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='Overtime Hours',bins=15)
plt.xlabel('Overtime Hours')
plt.title("Overtime Hours Table")
plt.ylabel("Frequency")
plt.show()
```

```python
Q1= df['Overtime Hours'].quantile(0.25)
Q3=df['Overtime Hours'].quantile(0.75)

IQR=Q3-Q1
upper=Q3 + 1.5*IQR
Lower=Q1 - 1.5*IQR

print(upper)
print(Lower)
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df['Overtime Hours'])
plt.title('Overtime Hours Distribution')
plt.xlabel('Overtime Hours')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(data=df,x='Overtime Hours')
plt.show()
```

```python
not_outlier_df=df[(df['Overtime Hours']>Lower)&(df['Overtime Hours']<upper)]
not_outlier_df
```

- **Most Employees do their Overtime Hours are 0 to 20** 

# BiVariate Analysis

```python
df.groupby(['Gender'])['Monthly Salary'].mean()
```

- Males and Females are equally Paid in this Company

```python
df.groupby(['Gender'])['Sick Leaves'].value_counts()
```

- Male Employees overall did not leave sick leaves but more men than women used their sick leaves 

```python
df.groupby(['Gender'])['Unpaid Leaves'].value_counts()
```

- Male Employees overall did not leave unpaid leaves but more men than women used their unpaid leaves 

```python
df.groupby(['Gender'])['Job Rate'].mean()
```

- Male and Female has the same Job rating 

```python
df.groupby(['Gender'])['Department'].describe()
```

```python
df.groupby(['Gender'])['Years'].value_counts()
```

- There is no Female with 5 years of work and with ratio, Female has more years of experience overall
